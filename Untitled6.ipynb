```python
from docx import Document
from docx.shared import Pt

# Create a new Document
doc = Document()
doc.add_heading('Elonpet Consulting Training List', 0)

# Add a table with 8 columns (Day, Name, Phone Number, Time-In, Time-Out, Signature)
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in days_of_week:
    doc.add_heading(day, level=1)
    table = doc.add_table(rows=1, cols=5)
    table.style = 'Table Grid'
    
    # Add the header row
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = 'Name'
    hdr_cells[1].text = 'Phone Number'
    hdr_cells[2].text = 'Time-In'
    hdr_cells[3].text = 'Time-Out'
    hdr_cells[4].text = 'Signature'
    
    # Set the font size of header
    for cell in hdr_cells:
        for paragraph in cell.paragraphs:
            for run in paragraph.runs:
                run.font.size = Pt(12)
    
    # Add blank rows for trainees (assuming 10 trainees for example)
    for _ in range(10):
        row_cells = table.add_row().cells
        for cell in row_cells:
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.size = Pt(12)

# Save the document
doc.save('Elonpet_TraineeList.docx')
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the data
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Books', 'Sports']
sales = [250, 150, 300, 100, 200]

# Convert the data to a NumPy array (optional, but good practice for larger datasets)
sales = np.array(sales)
# Create the pie chart
plt.figure(figsize=(8, 8))  # Optional: set the figure size
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Sales Distribution by Product Category')

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
csv_file_path = 'Documents/New Beverage launch Market Research (Responses) - Form responses 1.csv'
data = pd.read_csv(csv_file_path)
print(data.head())
preference_counts = data['Age'].value_counts()
labels = preference_counts.index
sizes = preference_counts.values
colors = plt.cm.Paired(np.linspace(0, 1, len(labels)))
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Beverage Preferences')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
csv_file_path = 'Documents/New Beverage launch Market Research (Responses) - Form responses 1.csv'
data = pd.read_csv(csv_file_path)
print(data.head())

# List of categorical columns to plot
categorical_columns = [
    'Age', 'Gender', 'How often do you consume Beverages?(like Soda, Juice e.t.c) per week',
    'What type of Beverages do you prefer?', 'What price range do you usually spend on a single beverage?',
    'Where do you usually buy your beverages? (Select all that apply)',
    'What time of day do you usually consume beverages? (Select all that apply)',
    'What type of new beverage product would you be most interested in? (Select all that apply)'
]

# Plot pie charts for each categorical column
for column in categorical_columns:
    counts = data[column].value_counts()
    labels = counts.index
    sizes = counts.values
    colors = plt.cm.Paired(np.linspace(0, 1, len(labels)))

    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Load the CSV file into a DataFrame
df = pd.read_csv('Documents/New Beverage launch Market Research (Responses) - Form responses 1.csv')
# Extract the column data
beverages = df['What type of Beverages do you prefer?'].dropna()
# Count the occurrences of each type of beverage
beverage_counts = beverages.value_counts()

# Plot the histogram
plt.figure(figsize=(12, 12))
plt.bar(beverage_counts.index, beverage_counts.values, color='blue')
plt.xlabel('Type of Beverages')
plt.ylabel('Frequency')
plt.title('Histogram of Preferred Beverages')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
git init

```
