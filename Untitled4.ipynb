```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = np.random.randn(1000)

# Plot histogram
plt.figure()
plt.hist(data, bins=30, alpha=0.6, color='blue')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.savefig('randomdata.png')  # Save histogram as PNG file

# Generate sample data for line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot line graph
plt.figure()
plt.plot(x, y, label='sin(x)', color='red')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.savefig('line_graph.png')  # Save line graph as PNG file

# Show the plots (optional)
plt.show()
```

```python
from docx import Document
from docx.shared import Pt

# Create a new Document
doc = Document()
doc.add_heading('WeGoHostU Web & App Services - Trainee Attendance List', 0)

# Add a table with 8 columns (Day, Name, Phone Number, Time-In, Time-Out, Signature)
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in days_of_week:
    doc.add_heading(day, level=1)
    table = doc.add_table(rows=1, cols=5)
    table.style = 'Table Grid'
    
    # Add the header row
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = 'Name'
    hdr_cells[1].text = 'Phone Number'
    hdr_cells[2].text = 'Time-In'
    hdr_cells[3].text = 'Time-Out'
    hdr_cells[4].text = 'Signature'
    
    # Set the font size of header
    for cell in hdr_cells:
        for paragraph in cell.paragraphs:
            for run in paragraph.runs:
                run.font.size = Pt(12)
    
    # Add blank rows for trainees (assuming 10 trainees for example)
    for _ in range(10):
        row_cells = table.add_row().cells
        for cell in row_cells:
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.size = Pt(12)

# Save the document
doc.save('NewSavedDocument.docx')
```
